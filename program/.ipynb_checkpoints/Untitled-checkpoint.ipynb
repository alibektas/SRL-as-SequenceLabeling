# Was ist neu?

1. Frühere Annahme : je komprimierter die Kodierung desto besser.
    1. Prädikate sind die wichtigsten Elemente im Satz
    1. Alles wird unter ihnen eingefügt
    1. **Problem:** Tags wie 01++6ARGM-LOC++9ARG1 kommen einmal vor.
2. Wie kann man Signalfrequenz steigern?
3. Indizes verursachen Probleme : ++3ARG0 and ++6ARG0 haben dieselbe Bedeutung jedoch unterschiedliche Verweise.
    1. Lsg : Wichtigeren Wörtern im Satz neue Zahlen vergeben. 
    2. Zugeordnete Zahlen dann kann man noch auf eine maximale Zahl beschränken. Informationsverlust.
4. \%50.4 aller Wörter haben keine Tags. Kann man diese sinnvoll benutzen?
    1. Vorschlag: Die Wörter zwischen zwei Verben V1 und V2 können für die Kodierung von Tags von V1 verwendet werden. Ähnelt BIO/IOB Tagging.
